**Assignment 14: Model Explainability**
Task:

Explain why your model predicts a transaction as Fraud or Non-Fraud.

 **What is Explainability?**

Explainability refers to the ability to understand and interpret how and why a machine learning model makes a particular prediction.

In fraud detection systems, explainability is essential because:

Decisions impact real users

Incorrect predictions can cause financial harm

Transparency builds trust

**Explainability: Why X → Y?**
What X and Y Represent

X: Input transaction features (Time, V1–V28, Amount)

Y: Model output

0 = Non-Fraud

1 = Fraud

The model learns patterns from historical transactions and predicts whether a new transaction is fraudulent based on how closely it matches known fraud behaviors.

**Example 1: Why X → Fraud (Y = 1)**

A transaction is classified as Fraud because:

The transaction amount is significantly higher than normal.

Multiple features (V-values) show abnormal patterns.

The overall behavior differs strongly from typical user activity.

Interpretation:
The model recognizes this combination as similar to past fraudulent transactions, leading to a fraud prediction.

**Ethical Considerations**

Fairness: Legitimate users should not be penalized.

Bias Awareness: Imbalanced data may affect predictions.

Transparency: Model decisions should be explainable.

Trust: Clear explanations increase user confidence.
