**Markdown Cell #2 – AI in Finance**

 AI in Finance

In financial services, AI is widely used for:

1. **Fraud Detection:**  
   - Detect unusual patterns in credit card transactions.  
   - Reduce false positives by combining supervised and unsupervised models.

2. **Risk Assessment:**  
   - Evaluate creditworthiness using historical data.  
   - Predict defaults and prevent financial loss.

3. **Customer Insights:**  
   - Segment users based on spending patterns.  
   - Provide personalized recommendations and alerts.

Our project applies AI models such as **ANN, LSTM, and clustering** to
identify potential fraud in credit card transactions, mirroring the
approaches used in real-world banking systems.
